# Emotion classification using the RAVDESS dataset

The Ryerson Audio-Visual Database of Emotional Speech and Song (RAVDESS) is licensed under CC BY-NA-SC 4.0. 

#### Construction and Validatio

Construction and validation of the RAVDESS is described in our paper: Livingstone SR, Russo FA (2018) The Ryerson Audio-Visual Database of Emotional Speech and Song (RAVDESS): A dynamic, multimodal set of facial and vocal expressions in North American English. PLoS ONE 13(5): e0196391. https://doi.org/10.1371/journal.pone.0196391.

The RAVDESS contains 7356 files. Each file was rated 10 times on emotional validity, intensity, and genuineness. Ratings were provided by 247 individuals who were characteristic of untrained adult research participants from North America. A further set of 72 participants provided test-retest data. High levels of emotional validity, interrater reliability, and test-retest intrarater reliability were reported. Validation data is open-access, and can be downloaded along with our paper from PLOS ONE.

#### Description

The dataset contains the complete set of 7356 RAVDESS files (total size: 24.8 GB). Each of the 24 actors consists of three modality formats: Audio-only (16bit, 48kHz .wav), Audio-Video (720p H.264, AAC 48kHz, .mp4), and Video-only (no sound).  Note, there are no song files for Actor_18.

#### File naming convention

Each of the 7356 RAVDESS files has a unique filename. The filename consists of a 7-part numerical identifier (e.g., 02-01-06-01-02-01-12.mp4). These identifiers define the stimulus characteristics:

#### Filename identifiers

- Modality (01 = full-AV, 02 = video-only, 03 = audio-only).
- Vocal channel (01 = speech, 02 = song).
- Emotion (01 = neutral, 02 = calm, 03 = happy, 04 = sad, 05 = angry, 06 = fearful, 07 = disgust, 08 = surprised).
- Emotional intensity (01 = normal, 02 = strong). NOTE: There is no strong intensity for the ‘neutral’ emotion.
- Statement (01 = “Kids are talking by the door”, 02 = “Dogs are sitting by the door”).
- Repetition (01 = 1st repetition, 02 = 2nd repetition).
- Actor (01 to 24. Odd numbered actors are male, even numbered actors are female).

Filename example: 02-01-06-01-02-01-12.mp4 

- Video-only (02)
- Speech (01)
- Fearful (06)
- Normal intensity (01)
- Statement “dogs” (02)
- 1st Repetition (01)
- 12th Actor (12)
- Female, as the actor ID number is even.

```python
# Feature extraction function
def extract_features(file_path):
    audio, sample_rate = librosa.load(file_path, res_type='kaiser_fast')
    mfccs = np.mean(librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=13), axis=1)
    chroma = np.mean(librosa.feature.chroma_stft(y=audio, sr=sample_rate), axis=1)
    mel = np.mean(librosa.feature.melspectrogram(y=audio, sr=sample_rate), axis=1)
    spectral_contrast = np.mean(librosa.feature.spectral_contrast(y=audio, sr=sample_rate), axis=1)
    return np.hstack((mfccs, chroma, mel, spectral_contrast))

# Extract features for the entire dataset
X = np.array([extract_features(path) for path in dataPath['Path']])
```

```python
import librosa
import numpy as np
import pandas as pd

# Initialize an empty DataFrame to store MFCC features
mfcc_data = pd.DataFrame()

# Emotions
emotion_labels = dataPath['Emotions'].unique()

# MFCC feature labels
mfcc_labels = [f'MFCC{str(i + 1)}' for i in range(13)]

# Iterate through emotions and calculate MFCC features
for emotion in emotion_labels:
    paths = dataPath[dataPath['Emotions'] == emotion]['Path']
    mfcc_features = []

    for path in paths:
        audio, sr = librosa.load(path)
        mfcc = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=13)
        mfcc_mean = np.mean(mfcc, axis=1)
        mfcc_features.append(mfcc_mean)

    # Convert the list to a NumPy array
    mfcc_features = np.array(mfcc_features)

    # Calculate the mean of MFCC features
    mfcc_mean = np.mean(mfcc_features, axis=0)

    # Add the mean MFCC features to the DataFrame
    mfcc_data[emotion] = mfcc_mean

# Set the row index as the MFCC feature labels
mfcc_data.index = mfcc_labels

# Display the table
print(mfcc_data)
```

**MFCC1:** (Zeroth coefficient): This coefficient represents the overall signal energy or power.

**MFCC2:** It typically represents the balance between high and low-frequency components in the audio signal.

**MFCC3:** Captures the rate of change of the spectral envelope of the signal.

**MFCC4:** Similar to MFCC3, it measures spectral envelope dynamics.

**MFCC5:** This feature focuses on the balance between the spectral peaks and valleys.

**MFCC6:** It accounts for variations in the rate of spectral envelope change.

**MFCC7:** Captures the spectral shape, particularly emphasizing high-frequency components.

**MFCC8:** Similar to MFCC7, it represents spectral shape dynamics.

**MFCC9:** Focuses on high-frequency spectral details and dynamics.

**MFCC10:** Captures spectral fine structure, such as spectral peaks.

**MFCC11:** It's related to high-frequency spectral variability and spectral details.

**MFCC12:** This coefficient is sensitive to the spectral transitions or variations.

**MFCC13:** It represents the overall spectral variation.
